```python
def insert_lines(img,seeding = 42):
    H,W,C = img.shape #height, width, channel
    np.random.seed(seeding)
    num_lines = np.random.randint(0,30)
    for i in range(num_lines):
        strength = np.random.randint(5,50)
        w_pos = np.random.randint(0,W)
        img[:,w_pos,:] = img[:,w_pos,:] // strength
    return img
    
```
